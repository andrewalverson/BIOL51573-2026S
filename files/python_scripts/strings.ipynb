### Store and print a variable

```python
# define some strings
first = 'beyonce'
last  = 'knowles'

# concatenate those strings
full = first + last

# print the strings
print(first)
print(last)
print(full)
print(first + last + "first")
print(first + ' ' + last + ' is the queen')
```

### Get the length of a string

```python
# the long (but readable) way
name_length = len(full)
print(name_length)

# a more concise way to do it, functions within functions
print(len(full))
```

### Print strings and numbers

```python
# print strings and integers
print("There are " + str(name_length) + " characters in Beyonce's name.")
print("There are", str(name_length), "characters in Beyonce's name")

# formatted print
print(f"There are {name_length} characters in Beyonce's name")
```

## Some operations on strings

```python
# methods to change the case of a string
print(first, last)
print(first.upper())
print(last.lower())

# `type()` returns what type of variable something is
print(type(first))

# make the name properly cased (play with this string to make sure our algorith is robust)
db_name = 'AndY'

# can give a literal string to `capitalize()`
print("andy".capitalize())
# ultimately our algorthim is: lowercase it first, then capitalize it: should work on any string this way
db_name_lower = db_name.lower()
db_name_cap = db_name_lower.capitalize()
print(db_name_cap)

# here we do it all in one statement
print(db_name.lower().capitalize())
```

### Problem #1: Write a program that asks for the user's name and age, then prints a personalized greeting using string formatting (`.format`).

For *Andy* and *32*, the output would be:

Hello, Andy! You are 32 years old.


```python
# code here
name = input('Enter your name: ')
age  = input('Enter your age: ')

name = name.lower().capitalize()

print('Hello, {}. You are {} years old'.format(name, age))
print(f'Hello, {name}. You are {age} years old')
```

### Problem #4: Tip Calculator

Ask the user for the bill amount (float) and tip percentage (integer). Calculate the tip and total, then print them neatly.

Enter the bill amount: `42.75`

Enter the tip percentage: `18`

Bill: `$42.75`

Tip (18%): `$7.70`

\--------------------

Total: `$50.45`


```python
# problem 4 solution
bill        = input('Enter the bill amount: ')
tip_percent = input('Enter the tip percentage: ')

bill = float(bill)
tip_percent = float(tip_percent)
tip_amount = bill*(tip_percent/100)
total = bill + tip_amount

print(f'Bill: {bill}')
print(f'Tip ({tip_percent}%): ${tip_amount:.2f}')
print('--------------------')
print(f'Total: ${total:.2f}')
```
